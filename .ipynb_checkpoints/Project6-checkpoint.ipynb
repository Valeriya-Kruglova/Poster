#### Сколько посетителей в день, неделю, месяц? <a name="посещения_период"></a>

```python
df['session_year'] = df['start_ts'].dt.year
df['session_month'] = df['start_ts'].dt.month
df['session_week'] = df['start_ts'].dt.week
df['session_date'] = df['start_ts'].dt.date
```

Добавили в датафрейм столбцы с годом, месяцем, неделей и полной датой. Рассчитаем средние значения посещений.

```python
user_day = df.groupby('session_date').agg({'uid':'nunique'})
DAU = user_day.mean()

user_week = df.groupby(['session_year', 'session_week']).agg({'uid':'nunique'})
WAU = user_week.mean()

user_month = df.groupby(['session_year', 'session_month']).agg({'uid':'nunique'})
MAU = user_month.mean()

print(int(DAU))
print(int(WAU))
print(int(MAU))
```

Среднее чило пользователей в день - 907 человек, в неделю чуть более 5700, а в месяц более 23000. Взглянем на регулярность посещения.

```python
sticky_MAU = DAU / MAU
sticky_WAU = DAU / WAU

print(sticky_WAU * 100)
print(sticky_MAU * 100)
```

Sticky Factor принимает почти минимальное значение, если рассматривать месячный период. Это говорит о нерегулярности посещения сайта, о том, что пользователя трудно удержать. Нормы этого показателя для информационного ресурса или сервиса для покупки билетов найти не удалось, но, в среднем, она составляет <a href="https://rb.ru/longread/product-metrics/">около 20%</a>

Посмотрим на поведение пользователей по дням.

```python
user_day.plot(figsize=(15, 5), grid=True, title='Количество посещений в зависимости от дня')
plt.xlabel('День')
plt.ylabel('Количество пользователей')
```

На данном графике видны резкие скачки активности и день, когда пользователей не было вообще. Скорее всего в конце марта 2018 сайт не был доступен, посмотрим когда был резкий всплеск активности.

```python
user_day.loc[user_day['uid'] > 3000]
```

К сожалению, не удалось найти весомой причины такой активности в этот день, может быть она вызвана акцией или рекламной компанией.

Рассмотрим 4 недели для того чтобы понять, существует ли зависимость от дня недели.<a name="пользователи_динамика"></a>

```python
user_day.head(28).plot(figsize=(15, 5), grid=True, title='Количество посещений в зависимости от дня недели')
plt.xlabel('День')
plt.ylabel('Количество пользователей')
```

Отчетливо видна зависимость посещений от дня недели: меньше всего посетителей в субботу и воскресение, а больше всего в четверг. Это вполне логично, скорее всего люди предпочитают искать мероприятия непосредственно перед выходными, а не во время отдыха.

```python
user_month.plot(figsize=(15, 5), grid=True, title='Количество посещений в зависимости от месяца')
plt.xlabel('Месяц')
plt.ylabel('Количество пользователей')
```

На графике видна сезонность посещений сайта. В теплое время года, люди предпочитают проводить время на свежем воздухе, зимой лучше посетить какое-то мероприятие.

Среднее количество посетителей в день - 907, в неделю чуть более 5700, а в месяц почти 23300. При этом Sticky Factor меньше 4%, что означает, что всего 4% пользователей заходит на ресурс повторно в течении месяца. Кроме того, посещаемость зависит от дня недели (в выходые - меньше всего посещений, в четверг - больше всего), покупатели предпочитают смотреть афишу и планировать свое время непосредственно перед выходными. Существует и сезонность: с октября по декабрь наблюдается всплеск активности, в теплые месяца наоборот - значительный спад.

#### Как часто люди возвращаются? <a name="польователи_возврат"></a>

Ранее уже был сделан вывод, что всего 4% пользователей возвращаются в течении месяца, но стоит посмотреть как изменяется это значение в зависимости от первой сессии. Проведем кагортный анализ.

```python
first_session_group = df.groupby('uid')['start_ts'].min()
first_session_group.name = 'first_session'
visits = df.merge(first_session_group, on='uid')
visits['first_session_month'] = visits['first_session'].astype('datetime64[M]')
visits['current_session_month'] = visits['start_ts'].astype('datetime64[M]')
visits.head()
```

Нашли для каждого посетителя дату первой сессии и выделили месяца текущей сессии и первой. Посмотрим на изменения в зависимости от возраста кагорты.

```python
visits['life_time'] = visits['current_session_month'] - visits['first_session_month']
visits['life_time'] = (visits['life_time'] / np.timedelta64(1, 'M')).round().astype('int')
visits['first_session_month'] = visits['first_session_month'].dt.strftime('%Y-%m')
visits_pivot = visits.pivot_table(index='first_session_month', columns='life_time', values='uid', aggfunc='nunique')
```

```python
plt.figure(figsize=(13, 9))
plt.title('Посещения сайта')
sns.heatmap(visits_pivot, annot=True, fmt='.0f', linewidths=1, linecolor='gray', vmin = 400, vmax=2500)
plt.xlabel('Порядковый месяц после первого посещения')
plt.ylabel('Месяц первого посещения')
```

Отчетливо видно, что после первого посещения, большая часть пользователей уже не возвращается на сайт и с каждым месяцом посетителей из каждой кагорты все меньше. Посмотрим на коэфициент удержания.

```python
cohorts_visits = visits.groupby(['first_session_month', 'life_time']).agg({'uid':'nunique'}).reset_index()
```

```python
initial_users_count_visits = cohorts_visits.query('life_time == 0')[['first_session_month','uid']]
initial_users_count_visits = initial_users_count_visits.rename(columns={'uid':'user_cohorts'})
cohorts_visits = cohorts_visits.merge(initial_users_count_visits,on='first_session_month') 
cohorts_visits['retention'] = cohorts_visits['uid'] / cohorts_visits['user_cohorts']
cohorts_visits.head()
```

```python
cohorts_visits['first_session_month'] = cohorts_visits['first_session_month'].astype('datetime64').dt.strftime('%Y-%m')
retention_pivot_visits = cohorts_visits.pivot_table(index='first_session_month',columns='life_time',values='retention',aggfunc='sum')
retention_pivot_visits.loc['mean'] = retention_pivot_visits.mean()
```

```python
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Коэфициент удержания')
sns.heatmap(retention_pivot_visits, annot=True, fmt='.1%', linewidths=1, linecolor='gray', vmin=0.02, vmax=0.1); 
plt.xlabel('Порядковый месяц после первого посещения')
plt.ylabel('Месяц первого посещения')
```

Рассчитаный ранее Sticky Factor и данная тепловая карта не противоречат друг другу, Retention Rate очень низкий и снижается с увеличением возроста когорты. Так, максимума он достигает для сентябрьской когорты в октябре, минимум приходится на май декабрьской кагорты, что подтверждает сезонность.

##### Вывод <a name="вывод4"></a>

В данном разделе подтвердили ранее полученные выводы о возвращаемости посететителей сайта. Вне зависимости от того, когда пользователь зашел на сайт, с каждым следующем месяцем вероятность того, что он больше его не посетит все больше. Так, в среднем, вероятность посещения сайта в следующий месяц - 6,5%, на следующий месяц уже 4,6% и чем больше возраст когорты, тем меньше этот процент.

#### Когда люди начинают покупать? <a name="первая_покупка"></a>

Для ответа на этот вопрос, нужно найти разницу между временем первой сессии и первой покупки. Дата и время первого посещения уже есть, найдем когда была совершена первая покупка и разницу между ними.

```python
first_orders = pd.DataFrame(orders.groupby('uid')['buy_ts'].min().reset_index())
first_session_orders = first_orders.join(first_session_group, on='uid')
first_session_orders['time_fs_fo'] = ((first_session_orders['buy_ts'] - first_session_orders['first_session']) / np.timedelta64(1,'m')).round().astype('int')
first_session_orders_pivot = first_session_orders.pivot_table(index='time_fs_fo', values='uid', aggfunc='count').reset_index()
```

```python
first_session_orders['time_fs_fo'].mode()
```

```python
first_session_orders['time_fs_fo'].max()
```

Так как максимальное значение слишком большое, а основная часть покупок совершается за минуту, для визуализации ограничим количество минут по оси x.

```python
first_session_orders_pivot.plot(x='time_fs_fo', y='uid', xlim=(0, 50), grid=True, title='Динамика времени покупки')
plt.xlabel('Время между первым посещением и первой покупкой (мин)')
plt.ylabel('Количество пользователей')
```

##### Вывод <a name="вывод5"></a>

Как правило, на покупку у пользователя уходит минута после первого посещения, дольше 30 минут - уже очень редко.

#### Средний чек <a name="средний_чек"></a>

```python
first_orders = first_orders.rename(columns={'buy_ts':'first_buy'})
orders_with_first = orders.merge(first_orders, on='uid')
orders_with_first['current_orders_month'] = orders_with_first['buy_ts'].astype('datetime64[M]')
orders_with_first['first_orders_month'] = orders_with_first['first_buy'].astype('datetime64[M]')
orders_cohort = orders_with_first.groupby(['first_orders_month', 'current_orders_month']).agg({'revenue':'sum', 'uid':'count'})
orders_cohort.head()
```

Сформировали когорты с данными о выручке и количестве покупок.

```python
orders_cohort['revenue_per_user'] = orders_cohort['revenue'] / orders_cohort['uid']
```

#### Сколько денег приносят (LTV)? <a name="LTV"></a>

Найдем количество новых покупателей за каждый месяц.

```python
first_orders['first_buy_month'] = first_orders['first_buy'].astype('datetime64[M]')
cohort_sizes = first_orders.groupby('first_buy_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_orders_month', 'n_buyers']
cohort_sizes.head()
```

Таблица `orders_with_first` уже содержит информацию о первом месяце покупки, текущем месяце и выручке. осталось сгруппировать, объединить и можно приступать к расчетам.

```python
cohorts = orders_with_first.groupby(['first_orders_month', 'current_orders_month']).agg({'revenue':'sum'}).reset_index()
cohorts.head()
```

```python
report = pd.merge(cohort_sizes, cohorts, on='first_orders_month')
report.head()
```

У нас нет данных о маржинальности, примем допущение, что выручка это и есть валовая прибыль. Найдем возрат когорт и LTV.

```python
report['age'] = ((report['current_orders_month'] - report['first_orders_month']) / np.timedelta64(1, 'M')).round().astype('int')
report['ltv'] = report['revenue'] / report['n_buyers']
report.head()
```

```python
output = report.pivot_table(
        index='first_orders_month', 
        columns='age', 
        values='ltv', 
        aggfunc='mean')
    
output.fillna('') 
```

```python
new_output = output.cumsum(axis=1).round(2)
new_output.fillna('')
```

```python
new_output.loc['mean'] = new_output.mean()
```

```python
plt.figure(figsize=(13, 9))
plt.title('LTV')
sns.heatmap(new_output, annot=True, fmt='.1f', linewidths=1, linecolor='gray', vmin=4, vmax=11) 
plt.xlabel('Порядковый месяц после первой покупки')
plt.ylabel('Месяц первой покупки')
```

##### Вывод <a name="вывод8"></a>

Самой прибыльной оказалась когорта снтября, но это вызано необычной покупкой в декабре, которую видно и в разделе с расчетом [среднего чека](#средний_чек). Фактор сезонности мешает проследить четкую динамику скорости накопления LTV у когорт, однако можно отметить, что с июня по сентябрь скорость накопления значительно выше следующих месяцев.

#### На сколько окупились расходы (ROI)? <a name="ROI"></a>

##### По времени <a name="ROI_время"></a>

```python
costs['month'] = costs['dt'].astype('datetime64[M]')
costs_monthly = costs.groupby('month')['costs'].sum()
costs_monthly.head()
```

Сгруппировали расходы по месяцам. Добавим информацию о расходах в существующую таблицу с расчетом LTV.

```python
report['first_orders_month'] = report['first_orders_month'].astype('datetime64')
```

```python
report_new = pd.merge(report, costs_monthly, left_on='first_orders_month', right_on='month')
report_new['cac'] = report_new['costs'] / report_new['n_buyers']
report_new.head()
```

```python
report_new['roi'] = report_new['ltv'] / report_new['cac']
output_new = report_new.pivot_table(
        index='first_orders_month', 
        columns='age', 
        values='roi', 
        aggfunc='mean')

output_cart = output_new.cumsum(axis=1).round(2) 
output_cart.head(12)
```

```python
plt.figure(figsize=(13, 9))
plt.title('ROI')
sns.heatmap(output_cart, annot=True, fmt='.2f', linewidths=1, linecolor='gray') 
plt.xlabel('Порядковый месяц после первой покупки')
plt.ylabel('Месяц первой покупки')
```

```python
output_new.cumsum(axis=1).mean(axis=0) 
```

Июньская когорта окупилась на 7-й месяц, сентябрьская на 4-й (но это только благодаря 1-й большой покупке). Все остальные не окупились. Посмотрим на линейном графике динамику затрат и выручки.

```python
orders['buy_month'] = orders['buy_ts'].astype('datetime64[M]')
dynamics_r = pd.DataFrame(orders.groupby('buy_month')['revenue'].sum().reset_index())
dynamics = pd.merge(dynamics_r, costs_monthly, left_on='buy_month', right_on='month', left_index=True)
dynamics.head()
dynamics[['revenue', 'costs']].plot()
plt.xlabel('Месяц')
plt.ylabel('Сумма')
```

И ROI и линейный график показывают, что затраты не окупаются. Основаня причина - низкий LTV, пользователи не возвращаются на ресурс и после высоких затрат на маркетинг, люди заходят на сайт, даже совершают покупку, но "пожизненая ценность" клиента остается низкой, потому что вероятность того, что он вернется в следующем месяце всего 4%.

##### По источникам

Определим источник пользователя и объединим с первыми покупками

```python
sources = df.sort_values(by = ['uid', 'start_ts']).groupby('uid').agg({'source_id':'first'})
payers = orders.groupby('uid').agg({'buy_ts':'min'}).merge(sources, on='uid').reset_index()
payers = payers.rename(columns = {'buy_ts':'first_buy'})
payers.head()
```

```python
ltv = payers.merge(orders, on='uid')
ltv['lifetime'] = ((ltv['buy_ts'] - ltv['first_buy']) / np.timedelta64(1, 'M')).round().astype('int')
display(ltv)
```

Посчитаем LTV

```python
orders_lifetime = ltv.pivot_table(index = 'source_id', columns = 'lifetime', values = 'revenue', aggfunc = 'count').fillna(0)
buyers_lifetime = ltv.pivot_table(index = 'source_id', columns = 'lifetime', values = 'uid', aggfunc = 'nunique').fillna(0)
ltv = ltv.pivot_table(index = 'source_id', columns = 'lifetime', values = 'revenue', aggfunc = 'sum').fillna(0)
ltv = ltv.cumsum(axis=1)
```

```python
cohort_sizes = payers.groupby('source_id').agg({'uid':'nunique'}).rename(columns={'uid':'cohort_size'})
ltv = ltv.merge(cohort_sizes, on='source_id')
ltv = ltv.div(ltv['cohort_size'], axis=0).drop(columns='cohort_size')
display(ltv)
```

Посчитаем CAC

```python
cac = costs.groupby('source_id').agg({'costs':'sum'}).reset_index().merge(cohort_sizes.reset_index(), on='source_id')
cac['cac'] = cac['costs'] / cac['cohort_size']
cac = cac.set_index('source_id')
display(cac)
```

Расчет ROMI

```python
romi = ltv.merge(cac[['cac']], on='source_id')
romi = romi.div(romi['cac'], axis=0).drop(columns='cac')
display(romi)
```

```python
plt.figure(figsize=(13, 9))
plt.title('ROMI')
sns.heatmap(romi, annot=True, fmt='.2f', linewidths=1, linecolor='gray') 
plt.xlabel('life_time')
plt.ylabel('Источник')
```
